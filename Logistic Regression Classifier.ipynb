```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,accuracy_score
```

```python
df= pd.read_csv("data.csv")
```

```python
df.head()
```

```python

df.info()
```

```python
df["Cuisines"]=df["Cuisines"].fillna("Unknown")  
```

```python
counts=df["Cuisines"].value_counts()
rare=counts[counts<10].index
df["Cuisines"]=df["Cuisines"].replace(rare,"other")
```

```python
x=TfidfVectorizer(stop_words="english",max_features=5000).fit_transform(df["Cuisines"])
y=df["Cuisines"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

```python
model=LogisticRegression(max_iter=2000,multi_class="multinomial",solver="lbfgs")
model.fit(x_train,y_train)
```

```python
y_predt=model.predict(x_test)
```

```python
accuracy_score(y_test,y_predt)
```

```python
classification_report(y_test,y_predt)
```
